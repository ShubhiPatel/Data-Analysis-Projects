```python
df.groupby('Job Title')['Salary'].mean().reset_index()
```

```python
avg_salary = df.groupby('Job Title')['Salary'].mean().reset_index()
```

```python
plt.figure(figsize = (10,6))
sns.barplot(x='Job Title', y='Salary', data= avg_salary, palette='viridis')
plt.xlabel('Job Title')
plt.ylabel('Average Salary ($)')
plt.title('Average Salary by Job Title')
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout() 
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Education Qualification', y='Salary', data=df)
plt.xlabel('Education Qualification')
plt.ylabel('Salary')
plt.title('Salary Distribution by Education Qualification')
plt.show()
```

```python
avg_salary = df.groupby('Education Qualification')['Salary'].mean().reset_index()


plt.figure(figsize=(7, 5))
sns.barplot(x='Education Qualification', y='Salary', data=avg_salary, palette='viridis')
plt.xlabel('Education Qualification')
plt.ylabel('Average Salary')
plt.title('Average Salary by Education Qualification')
plt.show()
```

```python
top_earners = df.loc[df.groupby('Job Title')['Salary'].idxmax().head()]
```

```python
plt.figure(figsize=(6,5))
sns.barplot(x='Salary', y='Job Title', data=top_earners, palette='viridis')
plt.title('Top Earners in Each Job')
plt.xlabel('Salary')
plt.ylabel('Job Title')
plt.show()
```
